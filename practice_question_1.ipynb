```python
import time
import matplotlib.pyplot as plt

# Interpolation Search Function
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and key >= arr[low] and key <= arr[high]:

        comparisons += 1

        if low == high:
            if arr[low] == key:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if arr[pos] == key:
            return pos, comparisons

        if arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparisons


# User Input
n = int(input("Enter number of elements: "))

# Uniformly Distributed Sorted Array
arr = list(range(n))

print("\nGenerated Array:")
print(arr[:20], "...") if n > 20 else print(arr)

key = int(input("\nEnter key to search: "))

# Measure Execution Time
start_time = time.perf_counter()

position, comparisons = interpolation_search(arr, key)

end_time = time.perf_counter()

execution_time = (end_time - start_time) * 1e6  # microseconds

# Display Results
print("\n----- Result -----")

if position != -1:
    print(f"Element found at index: {position}")
else:
    print("Element not found")

print(f"Number of Comparisons: {comparisons}")
print(f"Execution Time: {execution_time:.4f} microseconds")

print("\nComplexity Analysis")
print("Average Time Complexity : O(log log n)")
print("Best Case Time Complexity : O(1)")
print("Worst Case Time Complexity : O(n)")
print("Space Complexity : O(1)")

# -------------------------------
# Performance Analysis Graph
# -------------------------------

sizes = [1000, 5000, 10000, 50000, 100000]
times = []

for size in sizes:
    test_arr = list(range(size))
    test_key = size // 2

    start = time.perf_counter()
    interpolation_search(test_arr, test_key)
    end = time.perf_counter()

    times.append((end - start) * 1e6)

plt.figure(figsize=(8, 5))
plt.plot(sizes, times, marker='o')

plt.title("Interpolation Search Execution Time")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (Microseconds)")
plt.grid(True)

plt.show()
```
